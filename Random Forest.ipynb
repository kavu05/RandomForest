```python
#Importing the required libraries
import pandas as pd  
import re  
import nltk  
```

```python
# Load dataset'agr_en_train.csv'
df=pd.read_csv('agr_en_train.csv', names=['Text','label'])
print(df)
```

```python
#creating feature and labels using iloc method of pandas
features = df.iloc[:, 0].values  
labels = df.iloc[:, 1].values  
print(labels)
```

```python
features_processed = []


for sentence in range(0, len(features)):  
    # Remove all the special characters
    feature_processed = re.sub(r'\W', ' ', str(features[sentence]))
 
    
    # remove all single characters
    feature_processed= re.sub(r'\s+[a-zA-Z]\s+', ' ', feature_processed) 
    
    # Substituting multiple spaces with single space
    feature_processed = re.sub(r'\s+', ' ', feature_processed)

    # Converting to Lowercase
    feature_processed = feature_processed.lower()

    features_processed.append(feature_processed)
    print(features)
 
```

```python
#Feature Extraction/Vectorisation
from nltk.corpus import stopwords  
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=2500, min_df=7, max_df=0.8, stop_words=stopwords.words('english'))  
features_processed = vectorizer.fit_transform(features_processed).toarray()  
print(features_processed)
```

```python
#spliting the dataset into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(features_processed, labels, test_size=0.2) 
print(X_test)
```

```python
#RandomForest classifier
from sklearn.ensemble import RandomForestClassifier

text_classifier = RandomForestClassifier(n_estimators=100)  
text_classifier.fit(X_train, y_train) 
```

```python
predictions = text_classifier.predict(X_test) 
```

```python
predictions
```

```python
#Evaluation
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

print(confusion_matrix(y_test,predictions))  
print(classification_report(y_test,predictions))  
print(accuracy_score(y_test, predictions))  
```
